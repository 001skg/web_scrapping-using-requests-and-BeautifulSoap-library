```python
import requests
from bs4 import BeautifulSoup
 
 
# Making a GET request
result = requests.get('https://en.wikipedia.org/wiki/List_of_FIFA_World_Cup_finals')
 
# check status code for response received
# success code - 200
print(result)
 
# Parsing the HTML
soup = BeautifulSoup(result.content, 'html.parser')
print(soup.prettify())
```
